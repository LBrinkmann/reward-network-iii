```python
import pandas as pd
import seaborn as sns
import os
import json
```

```python
experiment_type = 'sim_3states_remote_50_v3'
```

```python
def load_trials(experiment_type):
    # Load trials from json
    trials_file = os.path.join('data', 'simulations', f'{experiment_type}.json')
    with open(trials_file, 'r') as f:
        trials = json.load(f)
    return trials
```

```python
trials = load_trials(experiment_type)
trials_df = pd.DataFrame(trials)
```

```python
sns.histplot(trials_df['response_time'])
```
